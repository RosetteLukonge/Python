```python
import numpy as np
my_array=np.random.uniform(5,10, size=(5,3))
my_array
```

```python
np.mean(my_array, axis=0)
```

```python
np.median(my_array, axis=0)
```

```python
np.std(my_array)
```

```python
n = 3 #number of nan values to insert into the array
position = np.random.choice(my_array.size, n, replace=False) #random position
my_array.ravel()[position] = np.nan #inserting nan values
print(my_array)
```

```python
url='https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
the_data_2d=np.genfromtxt(url,delimiter=',',dtype=float,usecols=[0,1,2,3])
the_data_2d
```

```python
the_data_2d_filter =[]
for item in the_data_2d:
    if item[2]>1.5 and item[0]<5.0:
        the_data_2d_filter.append(item)
the_data_2d_filter
```

```python
np.isnan(the_data_2d)
```

# Pandas Exercises

```python
import pandas as pd
data = {'Name': ['Mike','Lisa','George','Maria','Frank'],'Assignment 1': [7, 6.5, 4, 7,5], 'Assignment 2': [10, 9, 3, 9.5, 5], 'Assignment 3': [5.5, 8, 7, 4, 5]}
my_data_frame=pd.DataFrame(data, index=[1,2,3,4,5])
#my_data_frame=pd.DataFrame(data)
my_data_frame
```

```python
condition1=my_data_frame[my_data_frame['Assignment 1']<7]
condition1
```

```python
condition2=my_data_frame[(my_data_frame['Assignment 1']>6) & (my_data_frame['Assignment 2']>6) ]
condition2
```

```python
condition3=my_data_frame[(my_data_frame['Assignment 1']<5) | (my_data_frame['Assignment 2']<5) | (my_data_frame['Assignment 3']<5) ]
condition3
```

```python
sorted_dataFrame=my_data_frame.sort_values(['Name'], ascending=True)
sorted_dataFrame
```
